```python
import json
import requests
all_countries = json.loads(requests.get("https://restcountries.eu/rest/v1/all").text)

# [u'languages', u'nativeName', u'callingCodes',
#  u'gini', u'name', u'alpha3Code', u'area', u'demonym',
#  u'region', u'translations', u'alpha2Code', u'timezones',
#  u'subregion', u'capital', u'relevance', u'borders', u'topLevelDomain', 
#  u'currencies', u'latlng', u'altSpellings', u'population']
```

```python

country_names = {}
for country in all_countries:
    
    name = country['name']
    latlng = country['latlng']
    alt = country['altSpellings']
    demonym = country['demonym']
    native = country['nativeName']

    if latlng != []:
        if demonym != '':
            names = filter(None, alt)
            names.append(name)
            names.append(native)
            names.append(demonym)
            for n in names:
                country_names[n] = (name, latlng)

            
```
